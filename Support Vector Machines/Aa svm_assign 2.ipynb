```python
#importing libraries

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
```

```python
data1 = pd.read_csv('forestfires.csv')
data1
```

```python
label=LabelEncoder()
data1['size_category']=label.fit_transform(data1['size_category'])
```

```python
data1.drop(["month","day"],axis=1,inplace =True)
```

```python
y=data1.iloc[:,-1]
x=data1.drop('size_category',axis=1)
```

```python
# Standardization
def norm_func(i):
    x= (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
fires = norm_func(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,stratify = y)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
svc = SVC() 
params=[{'kernel':['rbf'],'gamma':[50,0.5,10,0.001],'C':[15,13,12,11,0.001]}]
gsv=GridSearchCV(svc,params,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_      # by this value we can find best , c and gamma to be used
```

```python
clf = SVC(kernel='rbf',C= 12, gamma =0.001 )
clf.fit(x_train , y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```
