```python
#importing libraries

import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
movies=pd.read_csv("my_movies (1).csv")
movies
```

```python
movies.drop(['V1','V2','V3','V4','V5'],axis=1,inplace=True)
```

```python
movies
```

```python
frequent_itemsets = apriori(movies, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.8)
rules
```

```python
rules.sort_values('lift',ascending = False)
```

```python
rules[rules.lift>1]
```

```python
import matplotlib.pyplot as plt
plt.scatter(rules['support'], rules['confidence'],c=rules['lift'],cmap= 'summer',alpha=0.75)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
cbar=plt.colorbar()
cbar.set_label('lift')
```
